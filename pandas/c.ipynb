```python
data_dict = {
    'data_x':[1, 2, 3, 4],
    'data_y':[1, 4, 9, 16]
}

plt.figure(figsize= (4,3))
plt.plot('data_x', 'data_y', data= data_dict)
# plt.xlabel('X', labelpad= 10, color = 'b', family = 'monospace', size = 20, weight = 'bold')
plt.xlabel('X', labelpad= 10, fontdict= {'color' : 'b', 
                                         'family' : 'monospace', 
                                         'size' : 20, 
                                         'weight' : 'bold'})

plt.ylabel('X^2')
plt.title("X vs X^2")
plt.show()
```

```python
plt.figure(figsize= (3,3))
plt.plot(np.arange(1,20), np.arange(1,20)**2, label = 'X^2')
plt.plot(np.arange(1,20), np.arange(1,20)**2.2, label = 'X^2.2')
plt.xlabel('X', loc = 'right')
plt.ylabel('Y', loc = 'top')
plt.title('Title')

plt.legend(ncol = 2, fontsize = 8, frameon = True, edgecolor = 'r')
plt.show()
```

```python
plt.figure(figsize= (3,3))
plt.plot(np.arange(1,20), np.arange(1,20)**2, label = 'X^2')
plt.plot(np.arange(1,20), np.arange(1,20)**3, label = 'X^3')

plt.xlabel('X', loc = 'right')
plt.ylabel('Y', loc = 'top')
# plt.xlim(0,10)
# plt.ylim(0,1000)
plt.axis([0, 10, 0, 1000])
plt.title('Title')
plt.legend(ncol = 2, fontsize = 8, frameon = True, edgecolor = 'r')

plt.show()
```

```python
## line type
plt.figure(figsize= (3,3))
plt.plot(np.arange(1,20), np.arange(1,20)**2, label = 'Dashed',
         linestyle = '--')
plt.plot(np.arange(1,20), np.arange(1,20)**3, label = 'Dotted',
         linestyle = ':')

plt.xlabel('X', loc = 'right')
plt.ylabel('Y', loc = 'top')

plt.axis([0, 10, 0, 1000])
plt.title('Title')
plt.legend(ncol = 2, fontsize = 8, frameon = True, edgecolor = 'r')

plt.show()
```

```python
plt.figure(figsize= (6, 4))
plt.plot([1, 2, 3], [4, 4, 4], linestyle = '-', marker = '.',
         color = 'r', label = 'Solid', linewidth = 2)
plt.plot([1, 2, 3], [3, 3, 3], linestyle = '--', marker = '+',
         color = 'b', label = 'Dashed', linewidth = 2)
plt.plot([1, 2, 3], [2, 2, 2], linestyle = ':', marker = 'o',
         color = 'c', label = 'Dotted')
plt.plot([1, 2, 3], [1, 1, 1], linestyle = '-.', marker = '*',
         color = 'm', label = 'Dash-dotted')
plt.ylim(0, 5)
plt.legend(ncol = 2, loc = 'upper center')
plt.show()
```
