```python
# R을 누르면 리턴 S를 누르면 저장

img = cv2.imread('./fig2/puppy.bmp', 0)
img_save = img.copy()

cv2.namedWindow('img', cv2.WINDOW_AUTOSIZE)
cv2.imshow('img', img)

while 1:
    key = cv2.waitKey()
    if key == ord('q') or key == 27:
        break

    elif key == ord('i'):
        img = 255 - img
        cv2.imshow('img', img)
        
    elif key == ord('c'):
        img = cv2.Canny(img, 50, 100) 
        cv2.imshow('img', img)
        
    elif key == ord('r'):
        img = img_save
        cv2.imshow('img', img)
    
    elif key == ord('s'):
        cv2.imwrite('output_image.jpg', img)
        
cv2.destroyAllWindows()
```

```python
# f누르면 좌우반전 c 엣지
f_flag = False
e_flag = False

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('video open failed')
    sys.exit()
    
while 1:
    ret, frame = cap.read()
    
    if not ret:
        print('frame read failed')
        break
    
    edge = cv2.Canny(frame, 50, 150)
    flipped = cv2.flip(frame, 1) # 좌우반전

    if f_flag:
        frame = flipped
    
    if e_flag:
        frame = edge

    key = cv2.waitKey(20)
    
    cv2.imshow('frame', frame)
    
    if key == ord('q') or key == 27:
        break
          
    elif key == ord('c'):
        e_flag = not e_flag
        cv2.imshow('frame', frame)
        
    elif key == ord('f'):
        f_flag = not f_flag

cv2.destroyAllWindows()
cap.release()
```

```python
def call_mouse(event, x, y, flags, param):
    global oldx, oldy
    
    if event == cv2.EVENT_LBUTTONDOWN:
        oldx, oldy = x, y
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if flags == cv2.EVENT_FLAG_LBUTTON:
            cv2.line(img, (oldx, oldy), (x, y), (0, 0, 255), 10, cv2.LINE_AA)
            cv2.imshow('img', img)
            oldx, oldy = x, y

img = np.ones((600, 600, 3), np.uint8)*255



cv2.namedWindow('img', cv2.WINDOW_AUTOSIZE)
cv2.setMouseCallback('img', call_mouse)

cv2.imshow('img', img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
def call_mouse(event, x, y, flags, param):
    global oldx, oldy
    
    if event == cv2.EVENT_LBUTTONDOWN:
        oldx, oldy = x, y
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if flags == cv2.EVENT_FLAG_LBUTTON:
            cv2.line(img, (oldx, oldy), (x, y), (0, 0, 255), 10, cv2.LINE_AA)
            cv2.imshow('img', img)
            oldx, oldy = x, y

img = np.ones((600, 600, 3), np.uint8)*255
img_save = img.copy()


cv2.namedWindow('img', cv2.WINDOW_AUTOSIZE)
cv2.setMouseCallback('img', call_mouse)

cv2.imshow('img', img)


while 1:
    key = cv2.waitKey()
    if key == ord('q') or key == 27:
        break
        
    elif key == ord('r'):
        img = img_save.copy()
        cv2.imshow('img', img)
    
    elif key == ord('s'):
        cv2.imwrite('image.jpg', img)
        
cv2.destroyAllWindows()
```
