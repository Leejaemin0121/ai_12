```python
img = cv2.imread('./fig/cat.bmp', cv2.IMREAD_GRAYSCALE)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # plt = rgb로 인식하기 때문에 변환

if img is None:
    print('이미지 리드 실패')
    sys.exit()

plt.imshow(img, 'gray')
plt.axis('off')
plt.show()
```

```python
img_bgr = cv2.imread('./fig/cat.bmp', cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

fig, ax = plt.subplots(1, 3, figsize = (12, 5))
ax[0].imshow(img_bgr), ax[0].axis('off')
ax[1].imshow(img_rgb), ax[1].axis('off')
ax[2].imshow(img_gray, cmap = 'gray'), ax[2].axis('off')
plt.show()
```

```python
img_files = os.listdir('./fig/images/')
file_paths = []

for i in img_files:
    file_path = './fig/images/' + i
    file_paths.append(file_path)
    
# display(file_path)

cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN, 
                      cv2.WINDOW_FULLSCREEN)

for i in file_paths:
    img = cv2.imread(i, cv2.IMREAD_COLOR)
    
    if img is None:
        continue
    
    cv2.imshow('scene', img)
    
    if cv2.waitKey(3000) == 27:
        break
    
cv2.destroyAllWindows()
```

```python
img_files = os.listdir('./fig/images/')
file_paths = []

for i in img_files:
    file_path = './fig/images/' + i
    file_paths.append(file_path)
    
# display(file_path)

cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN, 
                      cv2.WINDOW_FULLSCREEN)

idx = 0

while 1: 
    img = cv2.imread(file_paths[idx], cv2.IMREAD_COLOR)
        
    if img is None:
        continue
    
    cv2.imshow('scene', img)
    
    if cv2.waitKey(3000) == 27:            
        break

    idx += 1
    
    if idx >= len(file_paths):
        idx = 0
    
cv2.destroyAllWindows()
```
