```python
import sys
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
```

```python
src = cv2.imread('./fig2/cow.png')
dst = cv2.imread('./fig2/green.png')
src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

_, mask = cv2.threshold(src_gray, 240, 255, cv2.THRESH_BINARY_INV)

# size
h, w = src.shape[:2]
dst = cv2.resize(dst, (w, h), interpolation=cv2.INTER_AREA)

cv2.copyTo(src, mask, dst)

cv2.namedWindow('dst', cv2.WINDOW_NORMAL)

cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.imshow('mask', mask)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img_files = os.listdir('./fig/images/')
file_paths = []

for i in img_files:
    file_path = './fig/images/' + i
    file_paths.append(file_path)
    
# display(file_path)

cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN, 
                      cv2.WINDOW_FULLSCREEN)

idx = 0

while 1: 
    img = cv2.imread(file_paths[idx], cv2.IMREAD_COLOR)
        
    if img is None:
        continue
    
    cv2.imshow('scene', img)
    
    if cv2.waitKey(3000) == 27:            
        break

    idx += 1
    
    if idx >= len(file_paths):
        idx = 0
    
cv2.destroyAllWindows()
```

```python
img = np.ones((600, 1200, 3), np.uint8)*255

## 라인
cv2.line(img, (50, 80), (200,80), (0, 0, 255), 5, cv2.LINE_4)
cv2.line(img, (200, 80), (100,250), (0, 0, 255), 5, cv2.LINE_AA)
cv2.arrowedLine(img, (250, 50), (250,250), (0, 0, 255), 5, cv2.LINE_4)

## 사각형
# cv2.rectangle(img, (120, 300), (240,400), (0, 0, 255), 5, cv2.LINE_4)
cv2.rectangle(img, [120, 300, 120, 100], (0, 0, 255), -1, cv2.LINE_4)

## 원
cv2.circle(img, (800, 300), 100, (0, 0, 255), 5, cv2.LINE_AA)
## 타원
cv2.ellipse(img, (800, 300), (50, 100), 45, 0, 360, (255, 0, 0), 5, cv2.LINE_AA)

## 글자
cv2.putText(img, 'Hello openCV!', (600, 100), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 5, cv2.LINE_AA)

cv2.imshow('img', img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('video open failed')
    sys.exit()
    
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

out = cv2.VideoWriter('output.avi', fourcc, fps, (w, h))

while 1:
    ret, frame = cap.read()
    
    if not ret:
        print('frame read failed')
        break
    edge = cv2.Canny(frame, 50, 150)
    
    flipped_edge = cv2.flip(edge, 1) # 좌우반전

    cv2.putText(frame, 'Hello openCV!', (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 5, cv2.LINE_AA)
    
    cv2.imshow('frame', frame)
    cv2.imshow('edge', edge)
    cv2.imshow('flipped_edge', flipped_edge)
    
    out.write(frame)
    
    if cv2.waitKey(20) == 27:
        break

# cv2.waitKey()
cv2.destroyAllWindows()
cap.release()
out.release()

# print(cap.isOpened())
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('video open failed')
    sys.exit()
    
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

out = cv2.VideoWriter('output.avi', fourcc, fps, (w, h), isColor=False)

while 1:
    ret, frame = cap.read()
    
    if not ret:
        print('frame read failed')
        break
    edge = cv2.Canny(frame, 50, 150)
    
    flipped_edge = cv2.flip(edge, 1) # 좌우반전

    cv2.putText(frame, 'Hello openCV!', (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 5, cv2.LINE_AA)
    
    cv2.imshow('frame', frame)
    cv2.imshow('edge', edge)
    cv2.imshow('flipped_edge', flipped_edge)
    
    out.write(flipped_edge)
    
    if cv2.waitKey(20) == 27:
        break

# cv2.waitKey()
cv2.destroyAllWindows()
cap.release()
out.release()

# print(cap.isOpened())
```
