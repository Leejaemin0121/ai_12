```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

model = Sequential()    #  
model.add(Dense(1, input_dim=1, activation='linear'))   # layer 
model.compile(optimizer='sgd', loss='mse')  #   

#  
model.fit(x, y, epochs=2000)

# 
plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()
```

```python
hour=7
prediction = model.predict([hour])
print('%.f     %.2f.'%(hour, prediction))
```

```python
#   XOR  

w11 = np.array([-2, -2])
w12 = np.array([2, 2])
w2 = np.array([1, 1])
b1 = 3
b2 = -1
b3 = -1

#  
def MLP(x, w, b):
    y = np.sum(w * x) + b
    if y <= 0:
        return 0
    else:
        return 1
    
# NAND 
def NAND(x1, x2):
    return MLP(np.array([x1, x2]), w11, b1)

# OR
def OR(x1, x2):
    return MLP(np.array([x1, x2]), w12, b2)

# AND
def AND(x1, x2):
    return MLP(np.array([x1, x2]), w2, b3)

# XOR
def XOR(x1, x2):
    return AND(NAND(x1, x2), OR(x1, x2))


```

```python
# 

data_set = np.loadtxt('./data/ThoraricSurgery3.csv', delimiter=',')
X = data_set[:, 0:16]
y = data_set[:, -1]
```

```python
X.shape
```

```python
# MLP  
model_sgd = Sequential()
model_sgd.add(Dense(30, activation='relu', input_dim=(X.shape[1])))
model_sgd.add(Dense(10, activation='relu'))
model_sgd.add(Dense(1, activation='sigmoid'))

model_adam = Sequential()
model_adam.add(Dense(30, activation='relu', input_dim=(X.shape[1])))
model_adam.add(Dense(10, activation='relu'))
model_adam.add(Dense(1, activation='sigmoid'))

#    
model_sgd.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
history_sgd = model_sgd.fit(X, y, epochs=2000, batch_size=32, verbose=False)

model_adam.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history_adam = model_adam.fit(X, y, epochs=2000, verbose=False)
```

```python
history_sgd.history
history_adam.history['loss'][-1]
```

```python
# ThoraricSurgery3.csv    
predictions_sgd = model_sgd.predict(X)
predictions_adam = model_adam.predict(X)

#   
print("SGD  :")
for i in range(len(predictions_sgd)):
    print(f": {y[i]}, : {round(predictions_sgd[i][0])}")
```

```python
print("Adam  :")
for i in range(len(predictions_adam)):
    print(f": {y[i]}, : {round(predictions_adam[i][0])}")
```

```python
correct = 0
total = len(predictions_sgd)

for i in range(len(predictions_sgd)):
    if round(predictions_sgd[i][0]) == y[i]:
        correct += 1

accuracy = (correct / total) * 100
print(f": {accuracy:.2f}%")


correct = 0
total = len(predictions_adam)

for i in range(len(predictions_adam)):
    if round(predictions_adam[i][0]) == y[i]:
        correct += 1

accuracy = (correct / total) * 100
print(f": {accuracy:.2f}%")
```
