```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

model = Sequential()    # 모델 생성
model.add(Dense(1, input_dim=1, activation='linear'))   # layer 추가
model.compile(optimizer='sgd', loss='mse')  # 학습 컵파일 옵션

# 모델 학습
model.fit(x, y, epochs=2000)

# 그래프화
plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()
```

```python
hour=7
prediction = model.predict([hour])
print('%.f시간을 공부할 경우의 예상 점수는 %.2f점입니다.'%(hour, prediction))
```

```python
# 다중 선형 회귀 분석 -. 텐서플로

x = np.array([[2, 0], [4, 4], [6, 2], [8, 3]])
y = np.array([81, 93, 91, 97])

model = Sequential()    # 모델 생성
model.add(Dense(1, input_dim=2, activation='linear'))   # layer 추가
model.compile(optimizer='sgd', loss='mse')  # 학습 컵파일 옵션

# 모델 학습
model.fit(x, y, epochs=2000)

# 그래프화
```

```python
# 로지스틱 회귀분석

x = np.array([2, 4, 6, 8, 10, 12, 14])
y = np.array([0, 0, 0, 1, 1, 1, 1])

model = Sequential()    # 모델 생성
model.add(Dense(1, input_dim=1, activation='sigmoid'))   # layer 추가
model.compile(optimizer='sgd', loss='binary_crossentropy')  # 학습 컵파일 옵션

# 모델 학습
model.fit(x, y, epochs=5000)

# 그래프화
plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()
```

```python
hour=7
prediction = model.predict([hour])
print('%.f시간을 공부할 경우의 예상 점수는 %.2f점입니다.'%(hour, prediction * 100))
```

```python
# 다중 퍼셉트론으로 XOR 문제 해결

w11 = np.array([-2, -2])
w12 = np.array([2, 2])
w2 = np.array([1, 1])
b1 = 3
b2 = -1
b3 = -1

# 퍼셉트론 함수
def MLP(x, w, b):
    y = np.sum(w * x) + b
    if y <= 0:
        return 0
    else:
        return 1
    
# NAND 게이트
def NAND(x1, x2):
    return MLP(np.array([x1, x2]), w11, b1)

# OR게이트
def OR(x1, x2):
    return MLP(np.array([x1, x2]), w12, b2)

# AND게이트
def AND(x1, x2):
    return MLP(np.array([x1, x2]), w2, b3)

# XOR게이트
def XOR(x1, x2):
    return AND(NAND(x1, x2), OR(x1, x2))


```

```python
# 폐암 수술 후 1년 이내에 사망자와 생존자의 자료를 가지고 학습
# 1. HIDDEN LAYER는 2개 중 -> 처음 노드 수가 30개, 두번째가 10개
# 2. 적절한 활성함수 사용
# 3. 손실함수와 옵티마이저를 sgd, adam을 사용해서 결과 비교
```

```python
# 훈련데이터

data_set = np.loadtxt('./data/ThoraricSurgery3.csv', delimiter=',')
X = data_set[:, 0:16]
y = data_set[:, -1]
```

```python
X.shape
```

```python
# MLP 모델 구성
model_sgd = Sequential()
model_sgd.add(Dense(30, activation='relu', input_dim=(X.shape[1])))
model_sgd.add(Dense(10, activation='relu'))
model_sgd.add(Dense(1, activation='sigmoid'))

model_adam = Sequential()
model_adam.add(Dense(30, activation='relu', input_dim=(X.shape[1])))
model_adam.add(Dense(10, activation='relu'))
model_adam.add(Dense(1, activation='sigmoid'))

# 모델 컴파일 및 학습
model_sgd.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
history_sgd = model_sgd.fit(X, y, epochs=2000, batch_size=32, verbose=False)

model_adam.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history_adam = model_adam.fit(X, y, epochs=2000, verbose=False)
```

```python
history_sgd.history
history_adam.history['loss'][-1]
```

```python
# ThoraricSurgery3.csv 데이터에 대한 예측 수행
predictions_sgd = model_sgd.predict(X)
predictions_adam = model_adam.predict(X)

# 예측 결과 출력
print("SGD 예측 결과:")
for i in range(len(predictions_sgd)):
    print(f"실제값: {y[i]}, 예측값: {round(predictions_sgd[i][0])}")
```

```python
print("Adam 예측 결과:")
for i in range(len(predictions_adam)):
    print(f"실제값: {y[i]}, 예측값: {round(predictions_adam[i][0])}")
```

```python
correct = 0
total = len(predictions_sgd)

for i in range(len(predictions_sgd)):
    if round(predictions_sgd[i][0]) == y[i]:
        correct += 1

accuracy = (correct / total) * 100
print(f"정확도: {accuracy:.2f}%")


correct = 0
total = len(predictions_adam)

for i in range(len(predictions_adam)):
    if round(predictions_adam[i][0]) == y[i]:
        correct += 1

accuracy = (correct / total) * 100
print(f"정확도: {accuracy:.2f}%")
```
