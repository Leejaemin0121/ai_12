```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 훈련데이터
train = pd.read_csv('./data/train1.csv')
X = train.iloc[:, :-1].values
y = train.iloc[:, -1].values

# 테스트데이타
test = np.loadtxt('./data/test1.csv', delimiter=',')
test_target = np.loadtxt('./data/gender_submission.csv')
X_test = test[:]
y_test = test_target[:]

X_test.shape
X.shape
```

```python
# MLP 모델 구성
model_sgd = Sequential()
model_sgd.add(Dense(30, activation='relu', input_dim=(X.shape[1])))
model_sgd.add(Dense(10, activation='relu'))
model_sgd.add(Dense(1, activation='sigmoid'))

model_adam = Sequential()
model_adam.add(Dense(30, activation='relu', input_dim=(X.shape[1])))
model_adam.add(Dense(10, activation='relu'))
model_adam.add(Dense(1, activation='sigmoid'))

# 모델 컴파일 및 학습
model_sgd.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
history_sgd = model_sgd.fit(X, y, epochs=200, batch_size=32)

model_adam.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history_adam = model_adam.fit(X, y, epochs=200)
```

```python
predictions_sgd = model_sgd.predict(X_test)
predictions_adam = model_adam.predict(X_test)

# 예측 결과 출력
print("SGD 예측 결과:")
for i in range(len(predictions_sgd)):
    print(f"실제값: {y_test[i]}, 예측값: {round(predictions_sgd[i][0])}")
```

```python
print("Adam 예측 결과:")
for i in range(len(predictions_adam)):
    print(f"실제값: {y_test[i]}, 예측값: {round(predictions_adam[i][0])}")
```

```python
correct = 0
total = len(predictions_sgd)

for i in range(len(predictions_sgd)):
    if round(predictions_sgd[i][0]) == y_test[i]:
        correct += 1

accuracy = (correct / total) * 100
print(f"정확도: {accuracy:.2f}%")


correct = 0
total = len(predictions_adam)

for i in range(len(predictions_adam)):
    if round(predictions_adam[i][0]) == y_test[i]:
        correct += 1

accuracy = (correct / total) * 100
print(f"정확도: {accuracy:.2f}%")
```

```python
def calculate_accuracy(predictions, y_true):
    correct = sum(round(pred[0]) == true for pred, true in zip(predictions, y_true))
    total = len(predictions)
    accuracy = (correct / total) * 100
    return accuracy
```

```python
sgd_accuracy = calculate_accuracy(predictions_sgd, y_test)
print(f"SGD 모델의 정확도: {sgd_accuracy:.2f}%")

adam_accuracy = calculate_accuracy(predictions_adam, y_test)
print(f"Adam 모델의 정확도: {adam_accuracy:.2f}%")
```
